```python
class Sand():
    def __init__(self):
        self.sand=[]
        self.part1=0
        self.part2=0
        self.cnt=0
        self.part1_finish=1
        self.part2_finish=1
        
    def input_data(self,data):
        for i in range(len(data),2,-1):
            data[i-1]-=data[i-3]
        self.sand.append(data)

    def make_rock_map(self,data):
        y=data[2]+3
        self.start=y
        self.arr=[['.' for j in range(y*2+1)]for i in range(y)]
        for i in self.sand:
            x=i[0]-500+self.start
            y=i[1]
            for j in range(2,len(i),2):
                if int(i[j])==0:                 #y로 움직일때
                    dummy=int(i[j+1])
                    if dummy >0:
                        for y in range(y,y+dummy+1):
                            self.arr[y][x]='#'
                    else:
                        for y in range(y,y+dummy-1,-1):
                            self.arr[y][x]='#'
                else:
                    dummy=int(i[j])
                    if dummy >0:
                        for x in range(x,x+dummy+1):
                            self.arr[y][x]='#'
                    else:
                        for x in range(x,x+dummy-1,-1):
                            self.arr[y][x]='#'

        for i in range(len(self.arr[0])):
            self.arr[self.start-1][i]='#'

    def part2_input_sand(self):
        while self.part2_finish:
            self.move_sand2()

    def move_sand2(self):
        x=self.start
        y=0
        while self.check2(y):
            if self.arr[y+1][x]=='.':
                y+=1
            elif self.arr[y+1][x-1]=='.':
                y+=1
                x-=1
            elif self.arr[y+1][x+1]=='.':
                y+=1
                x+=1
            else:
                self.arr[y][x]='O'
                self.cnt+=1
                break
        
    def check2(self,y):
        if self.arr[0][self.start]=='O':
            self.part2_finish=0
            self.part2=self.cnt
            return False
        if self.part1_finish:
            if y==len(self.arr)-3:
                self.part1=self.cnt
                self.part1_finish=0
            return True
        else:
            return True

    def print_part2_count(self):
        return self.part2

    def print_part1_count(self):
        return self.part1
```

```python
with open("C:/Users/ckehd/Documents/GitHub/adventofcode_2022/DK/14.txt","r") as f:
    data=f.readlines()

max_min=[0,1000,0] #x_max,x_min,y_max\
idx=0
sand=Sand()

for i in data:
    sand_data=[]
    i=''.join(i).split()

    for j in range(0,len(i),2):
        dummy=i[j].split(',')
        x=int(dummy[0])
        y=int(dummy[1])
        if x>max_min[0]:
            max_min[0]=x
        if x<max_min[1]:
            max_min[1]=x
        if y>max_min[2]:
            max_min[2]=y
        sand_data.append(x)
        sand_data.append(y)
    sand.input_data(sand_data)
    
sand.make_rock_map(max_min)
#sand.input_sand()
#sand.print_rock_map()
sand.part2_input_sand()
print(sand.print_part2_count())
print(sand.print_part1_count())
#print(sand.print_rock_map())
```
