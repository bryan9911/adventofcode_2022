```python
class Cave:
    
    def __init__(self,rocks,data):
        self.rocks=rocks
        self.tunnel=[['.' for i in range(5)]for j in range(7)]
        self.input_data=data
        self.tunnel_hight=-1
        self.move_cnt=0
        self.rock_stop=1
        #self.starting_point=[2,len(self.tunnel[0]-1)]
        
    def start(self,i):
        fall_rock=self.rocks[i%5]
        starting_point=[2,self.tunnel_hight+4] #len(self.tunnel[0])-2
        rock_map=[[fall_rock[j][k]+starting_point[k] for k in range(2)] for j in range((len(fall_rock)))]
        while self.rock_stop:
            rock_map=(self.move_rock(rock_map,self.input_data[self.move_cnt%len(self.input_data)]))
            if self.move_cnt%len(self.input_data)==0:
                print("초기화")
            self.move_cnt+=1
            rock_map=self.fall_rock(rock_map)
        self.rock_stop=1
        
    def move_rock(self,rock_map,moving):
        if moving=='<':
            moving_idx=[-1,0]
            rock=[[rock_map[i][j]+moving_idx[j] for j in range(2)] for i in range((len(rock_map)))]
            if self.check_tunnel_move(rock):
                return rock
            else:
                return rock_map
        else:
            moving_idx=[1,0]
            rock=[[rock_map[i][j]+moving_idx[j] for j in range(2)] for i in range((len(rock_map)))]
            if self.check_tunnel_move(rock):
                return rock
            else:
                return rock_map
        
    def fall_rock(self,rock_map):
        moving_idx=[0,-1]
        rock=[[rock_map[i][j]+moving_idx[j] for j in range(2)] for i in range((len(rock_map)))]
        if self.check_tunnel_fall(rock):
            return rock
        else:
            for i in rock_map:
                self.tunnel[i[0]][i[1]]='@'
            for i in rock_map:
                if i[1]>self.tunnel_hight:
                    dummy=i[1]-self.tunnel_hight+3
                    self.plus_tunnel(dummy)
                    self.tunnel_hight=i[1]
                
    def check_tunnel_move(self,rock):
        for i in range(len(rock)):
            if int(rock[i][0])<0 or int(rock[i][0])>=7:
                return False
            if self.tunnel[rock[i][0]][rock[i][1]]!='.':
                return False
        return True
    
    def check_tunnel_fall(self,rock):
        for i in rock:
            if int(i[1])<0:
                self.rock_stop=0
                return False
            if self.tunnel[i[0]][i[1]]!='.':
                self.rock_stop=0
                return False
        return True
    
    def plus_tunnel(self,dummy):
        for i in range(len(self.tunnel)):
            for j in range(dummy):
                self.tunnel[i].append('.')
                
    def print_tunnel(self):
        for i in self.tunnel:
            print(*i)
    
    def print_tunnel_hight(self):
        return self.tunnel_hight
```
