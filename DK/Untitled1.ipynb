```python
class Sand():
    def __init__(self):
        self.sand=[]
        self.part1=0
        self.part2=0
        self.cnt=0
        self.part1_finish=1
        self.part2_finish=1
        
    def input_data(self,data):
        self.sand.append(data)
        
    def print_data(self):
        return self.sand

    def make_rock_map(self,data):
        self.start=500-data[1]+1
        self.arr=[['.' for j in range(data[0]-data[1]+3)] for i in range(data[2]+3)]
        for i in self.sand:
            x=i[0]-data[1]+1
            y=i[1]
            for j in range(2,len(i),2):
                i[j]-=(data[1]-1)
                if i[j]-x!=0:
#                    print(i[j])
                    x_dummy=i[j]
                    if x_dummy<x:
                        x,x_dummy=x_dummy,x
                    for k in range(x,x_dummy+1):
                        self.arr[y][k]='#'
                    x=i[j]
                else:
                    y_dummy=i[j+1]
                    if y_dummy<y:
                        y,y_dummy=y_dummy,y
                    for k in range(y,y_dummy+1):
#                        print(k,x)
                        self.arr[k][x]='#'
                    y=i[j+1]

#               self.print_rock_map()
        y=data[2]+2
        for i in range(len(self.arr[0])):
            self.arr[y][i]='#'

    def print_rock_map(self):
        for i in self.arr:
            print(i)

    def input_sand(self):
        while self.part1_finish:
            self.move_sand()
                    
    def move_sand(self):
        x=self.start
        y=0
        while self.check(y):
            if self.arr[y+1][x]=='.':
                y+=1
            elif self.arr[y+1][x-1]=='.':
                y+=1
                x-=1
            elif self.arr[y+1][x+1]=='.':
                y+=1
                x+=1
            else:
                self.arr[y][x]='O'
                self.part1+=1
                break


    def check(self,y):
        if y==len(self.arr)-3:
            self.part1_finish=0
            return False
        else:
            return True


    def part2_input_sand(self):
        while self.part2_finish:
            self.move_sand2()

    def move_sand2(self):
        x=self.start
        y=0
        while self.check2():
            if self.range_check(y+1,x):
                y+=1
            elif self.range_check(y+1,x-1):
                y+=1
                x-=1
            elif self.range_check(y+1,x+1):
                y+=1
                x+=1
            else:
                self.arr[y][x]='O'
                self.part2+=1
                break

    def range_check(self,y,x):
        if 0<=y<len(self.arr) and 0<=x<len(self.arr[0]):
            if self.arr[y][x]=='.':
                return True
        else:
            return False
        
    def check2(self):
        if self.arr[0][self.start]=='O':
            self.part2_finish=0
            return False
        else:
            return True

    def print_part2_count(self):
        return self.part2

    def print_part1_count(self):
        return self.part1
```
